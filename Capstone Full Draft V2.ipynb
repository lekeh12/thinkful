```python
#loading packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sp
from scipy.stats import ttest_ind
from scipy.stats import linregress
from scipy import stats
%matplotlib inline
```

```python
#datasets
df_socmed = pd.read_csv('hookit_data_sept17.csv', index_col=0, header=0, parse_dates=True)
df_playerdata = pd.read_csv('player_data_Sept2017.csv', index_col=0, parse_dates=True)
df_clubrev = pd.read_csv('Club Revenue Deloitte.csv', parse_dates=True)
```

```python
print(''' Ronaldo, Messi, and Neymar are the most valuable footballers in the world.  They also have the most followers on social media of any football players. Ronaldo has a staggering 295 million followers.  He has over 122 million on facebook. As a part of my general pursuit for the variables that drive player value, I wanted to look for variables that relate to number of followers on social media.  I used a dataset from Hookit, which is a company that measures and analyzes social media engagement.  I pulled the September 2017 data for football players with the most followers on social media.  They only look at facebook, twitter, and instagram.  I also used the September 2017 Transfermarkt data.  Transfermarkt keeps track of market value of players and also logs biographical data including information on injuries and performance.  Finally I included Deloitte's Football Money League report which is a review of the financial performance of major football clubs.  The data I used is from the 2016 report, and I mainly used the revenue data.''')
```

```python
print('''It's important to note the limitations on the data.  The Hookit dataset only has data from players on the top 100 list for social media. One thing that is a huge blindspot is the social media numbers from the Chinese and Russian networks.  
Players who are inactive are not included.   Transfermarkt has data available for active players only.  There are several like David Beckham who have powerful brands and who are still involved in the industry as coaches or managers such as Zinadine Zidane.''')
```
