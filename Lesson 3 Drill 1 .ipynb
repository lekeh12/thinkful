```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#wald distribution used for non negative positively skewed data
#tail decreases slowly compared to normal dist
#its like the gamma but wald is easier if there areextremely large values
#also called inverse gaussian
#could be used for survival analysis

waldy = np.random.wald(15, 200, 100)

plt.hist(waldy)

plt.axvline(waldy.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(waldy.mean() + waldy.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(waldy.mean() - waldy.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Rayleigh dist 
#ex to measure lifetime when object life depends on age like battery
#ex to model multiple paths of dense scattered signals reaching a receiver
#ex wave heights
rayleighly = np.random.rayleigh(3, 100)

plt.hist(rayleighly)

plt.axvline(rayleighly.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(rayleighly.mean() + rayleighly.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(rayleighly.mean() - rayleighly.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Gumbel dist 
#ex used for modelling floods
#characterized by fat tail, fatter tail than gaussian
#its an extreme value distribution
gumbelly = np.random.gumbel(2, .6, 100)

plt.hist(gumbelly)

plt.axvline(gumbelly.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(gumbelly.mean() + gumbelly.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(gumbelly.mean() - gumbelly.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#beta dist
#represent outcomes for percentages
#used for hypothesis testing as in can the results from this experiment be replicated
#can take values between 0 and 1
betab = np.random.beta(.4, .3, 100)

plt.hist(betab)

plt.axvline(betab.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(betab.mean() + betab.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(betab.mean() - betab.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#triangular distribution
#good when we have maximum and minumum maybe no mean or standard dev
#ex sum of two rolls of dice upper limit and lower limit clearly defined
#ex modelling weekly sales
#can be used to avoid extreme values

triangularrrr = np.random.triangular(10, 60, 110, 100)

plt.hist(triangularrrr)

plt.axvline(triangularrrr.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(triangularrrr.mean() + triangularrrr.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(triangularrrr.mean() - triangularrrr.std(), color='c', linestyle='dashed', linewidth=2)


plt.show()
```

```python
#exponential distribution
#represents a processin which events 
#occur continuously and independently at a constant average rate

exponentially = np.random.exponential(10, 100)

plt.hist(exponentially)

plt.axvline(exponentially.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(exponentially.mean() + exponentially.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(exponentially.mean() - exponentially.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
var1 = np.random.normal(5, .5, 100)
var2 = np.random.normal(10, 1, 100)
var3 = var1 + var2

plt.hist(var3)

plt.axvline(var3.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(var3.mean() + var3.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(var3.mean() - var3.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```
